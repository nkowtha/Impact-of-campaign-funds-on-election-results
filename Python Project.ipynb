```python
def process_missing_data(data, threshold):
    #calculate all columns with missing values percentage greater than the threshold
    missing_percentage = (df.isna().sum()/len(df))*100
    
    #check for threshold condition
    missing_filtered = missing_percentage[missing_percentage >= threshold] 
    
    #select the candidate columns to a list
    drop_cols = missing_filtered.index.tolist()
    
    # drop the columns in the data
    df.drop(columns=drop_cols, inplace=True)
    
    return df.shape
process_missing_data(df,90)
```
